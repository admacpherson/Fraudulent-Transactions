```python
# Imports
import kagglehub
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Load Kaggle dataset
def load_kaggle_csv(dataset, filename):
    path = Path(kagglehub.dataset_download(dataset))
    return pd.read_csv(path / filename)

df = load_kaggle_csv(
    "chitwanmanchanda/fraudulent-transactions-data",
    "Fraud.csv"
)
```

```python
# Concat origin and destination users
all_customers = pd.concat([df['nameOrig'], df['nameDest']])
```

```python
# Count repeat users
customer_counts = all_customers.value_counts()
num_unique_customers = customer_counts.size
num_repeat_customers = (customer_counts > 1).sum()
repeat_percentage = 100 * num_repeat_customers / num_unique_customers
print(f"Repeat customers comprise {repeat_percentage:.2f}% of the transactions")
```

```python
customer_merchant_count = all_customers.str[0].value_counts()
customer_merchant_count
```

```python
merchant_pct = customer_merchant_count[1] / (customer_merchant_count[0] + customer_merchant_count[1])
print(f"Merchants comprise {100*merchant_pct:.1f}% of the transactions")
```
